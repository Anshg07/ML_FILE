### PRGRAM FOR LINEAR REGRESSION IMPLEMENTATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
def simple_test_data():
    return pd.read_csv("./LR_test_train/test.csv")
```

```python
def simple_train_data():
    return pd.read_csv("./LR_test_train/train.csv")
```

```python
train_set = simple_train_data()
test_set = simple_test_data()
```

```python
train_set.head(), train_set.describe()
```

```python
test_set.head(), test_set.describe()
```

```python
train_set[train_set.y.isnull()]
```

```python
train_set.dropna(inplace=True)
```

```python
train_set[train_set.y.isnull()]
```

```python
test_set.isnull().sum()
```

```python
train_set.describe()
```

VISUALIZATION

try to visualize the data to get insights

```python
train_set.plot(kind='scatter', x='x', y='y', label="Train", alpha=0.5)
plt.show()
```

```python
test_set.plot(kind='scatter', x='x', y='y', label="Test", alpha=0.5)
plt.show()
```

Call the Linear Regression model

```python
model = LinearRegression()
```

```python
train_set_x = train_set['x'].values.reshape(-1, 1)
train_set_y = train_set['y']
```

```python
model.fit(train_set_x, train_set_y)
```

```python
print("Coefficient: ", model.coef_)
print("Intercept: ", model.intercept_.round(5))
```

Predicting the training set first

```python
predicted_train_set = pd.DataFrame(model.predict(train_set_x))
predicted_train_set.head()
```

To evaluate the performance of model we should calculate error/loss. Here I am using the root mean squared error.

```python
train_rmse = np.sqrt(mean_squared_error(train_set_y, predicted_train_set, squared=False))
print('RMSE: ', train_rmse)
```

The error is very low which means the predicted values are very much accurate

Now predicting the test set and values

```python
test_set_x = test_set['x'].values.reshape(-1, 1)
test_set_y = test_set['y']
```

```python
predicted_test_set = pd.DataFrame(model.predict(test_set_x))
```

```python
predicted_test_set.head()
```

Finding loss using RMSE.

```python
test_rmse = np.sqrt(mean_squared_error(test_set_y, predicted_test_set, squared=False))
print('RMSE: ', test_rmse)
```

Lets try to predict a value for 12.0 using this model.


```python
custom_predict = model.predict([[12.0]])
print('Predicted Value: ', custom_predict[0])
```

Calculating the value using the equation for Linear Regression.

y=β0+β1×X
 
Coefficient: 1.00065638 Intercept: -0.10727

y=−0.10727+1.00065638∗12.0=11.9006
 
"This is how the prediction is calculated."

There is an inbuild function 'score()' to find the accuracy ranging from 1.0 to 0.0, with 1.0 being most accurate.

```python
score_test = model.score(test_set_x, test_set_y)
print('Test Score: ',score_test)
```

```python
print('The LR model has accuracy of {} %'.format(round(score_test*100,2)))
```

Valuation through Plots

```python
plt.scatter(train_set_x, train_set_y, c='blue',alpha=0.7, label='Train')
plt.scatter(test_set_x, test_set_y, c='crimson', alpha=0.7, label='Test')
plt.legend()
plt.title('Train and Test Plot')
plt.show()
```

```python
plt.scatter(test_set_x, test_set_y, c='g', alpha=0.7, label='Test')
plt.plot(test_set_x, predicted_test_set, c='aquamarine', label='Predicted')
plt.legend()
plt.title('Model Predict Plot')
plt.show()
```
